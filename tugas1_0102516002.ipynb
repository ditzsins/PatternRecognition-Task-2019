```python
from scipy.stats import multivariate_normal

fig, ax = plt.subplots(figsize=(8,8))
mpg = np.linspace(df.mpg.min(), df.mpg.max(), df.mpg.count())
acc = np.linspace(df.acceleration.min(), df.acceleration.max(), df.acceleration.count())

for kelas, c in zip(['mpg', 'acceleration'], cmaps):
    mvn = multivariate_normal([df.mpg.mean(), df.acceleration.mean()], cov)

    m, a= np.meshgrid(mpg, acc)

    pos = np.empty(x1.shape + (2,))
    pos[:,:,0] = m
    pos[:,:,1] = a

    z = mvn.pdf(pos)
    ax.contour(m, a, z, alpha=1, colors='c')
    ax.scatter(df.mpg, df.acceleration,alpha=0.2, c='c')
    ax.set_xlabel('mpg')
    ax.set_ylabel('acceleration')
```

## 3. Distribusi Beta (10 poin)

Dua orang petinju, MP dan ODLH, akan bertemu satu masing-masing dalam sebuah pertandingan amal. MP telah bertanding 70 kali dan memenangkan 61 pertandingan, sedangkan ODLH menang 39 kali dari 45 pertandingannya. Asumsikan bahwa Anda tidak tahu apapun tentang kemampuan kedua petinju ini awalnya, lalu Anda ingin memodelkan peluang MP akan menang dari ODLH.

### Soal 3.1 (2 poin)

Jika Anda memodelkan kemampuan (*skill*) MP dan ODLH sebagai distribusi beta, berapa nilai parameter dari masing-masing model?

```python
aMP = 1 + 61 / 1
bMP = 1 + 70 - 61/1
aODLH = 1 + 39 / 1
bODLH = 1 + 45 - 39/1
print ('Alpha MP =' ,aMP)
print ('Beta MP =' ,bMP)
print ('Alpha ODLH =' ,aODLH)
print ('Beta ODLH =' ,bODLH)
```

Dengan mengasumsikan nilai menang = 1, dan kalah = 0, karena 0 >= X >= 1 

### Soal 3.2 (2 poin)

Gambarkan distribusi dari "skill" kedua petinju.

```python
from scipy.stats import beta
rv = beta(aMP, bMP)
x = np.linspace(beta.ppf(0.01, aMP, bMP), beta.pdf(0.99, aMP, bMP), 100)
fig, ax = plt.subplots(1,1)
ax.plot(x, rv.pdf(x),'c', lw=2, label='beta pdf MP')

rv2 = beta(aMP, bMP)
x2 = np.linspace(beta.ppf(0.01, aODLH, bODLH), beta.pdf(0.99, aODLH, bODLH), 100)
ax.plot(x2, rv2.pdf(x),'m', lw=2, label='beta pdf ODLH')
plt.legend()
plt.show()
```

### Soal 3.3.a (4 poin)

Berapa nilai ekspektasi dari variansi dari kedua model?

```python
print ('Nilai ekspektasi dari petinju MP = ', aMP / (aMP + bMP))
print ('Nilai ekspektasi dari petinju ODLH = ', aODLH / (aODLH + bODLH))
print ('Nilai variansi dari petinju MP = ',(aMP*bMP)/((aMP + bMP)**2)*(aMP + bMP + 1))
print ('Nilai variansi dari petinju ODLH = ',(aODLH*bODLH)/((aODLH + bODLH)**2)*(aODLH + bODLH + 1))
```
